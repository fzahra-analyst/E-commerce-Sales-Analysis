1) Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

2) Load Final Dataset (orders_full.csv)

```python
orders_full = pd.read_csv("../Data/orders_full.csv")
orders_full.head()
```

3) GEO BASIC ANALYSIS

```python
# Group by state
state_summary = orders_full.groupby('customer_state').agg(
    total_orders=('price', 'count'),   # Number of orders
    total_sales=('price', 'sum')       # Total sales
).sort_values(by='total_sales', ascending=False)  # Sort by sales from highest to lowest
# Show the top 10 states
state_summary.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select the top 10 states by sales
top_states = state_summary.head(10)

# Set up the chart/figure
fig, ax1 = plt.subplots(figsize=(12,6))

# Draw total sales as a bar chart
sns.barplot(
    x=top_states.index, 
    y=top_states['total_sales'], 
    color='skyblue', 
    ax=ax1
)
ax1.set_ylabel('Total Sales ($)', color='blue')
ax1.set_xlabel('State')
ax1.set_title('Top 10 States: Total Sales & Number of Orders')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)

# Create a secondary axis for the number of orders
ax2 = ax1.twinx()
sns.lineplot(
    x=top_states.index, 
    y=top_states['total_orders'], 
    color='red', 
    marker='o', 
    ax=ax2
)
ax2.set_ylabel('Number of Orders', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.tight_layout()
plt.show()
```

ðŸ“ 3.2 Average Delivery Time per State

```python
delivery_state = orders_full.groupby("customer_state")["delivery_time_days"].mean().reset_index()
delivery_state = delivery_state.sort_values("delivery_time_days")

plt.figure(figsize=(10,6))
sns.barplot(data=delivery_state, x="customer_state", y="delivery_time_days")
plt.title("Average Delivery Time (Days) per State")
plt.xlabel("State")
plt.ylabel("Delivery Time (days)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

ðŸ“ 3.3 Average Freight Cost per State

```python
freight_state = orders_full.groupby("customer_state")["freight_value"].mean().reset_index()
freight_state = freight_state.sort_values("freight_value")

plt.figure(figsize=(10,6))
sns.barplot(data=freight_state, x="customer_state", y="freight_value")
plt.title("Average Freight Cost per State")
plt.xlabel("State")
plt.ylabel("Freight Value (BRL)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

ðŸ“ 4) CITY-LEVEL ANALYSIS

4.1 Top 20 Cities by Number of Orders

```python
city_orders = orders_full["customer_city"].value_counts().head(20).reset_index()
city_orders.columns = ["city", "total_orders"]

plt.figure(figsize=(10,6))
sns.barplot(data=city_orders, x="total_orders", y="city")
plt.title("Top 20 Cities by Number of Orders")
plt.xlabel("Orders")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

4.2 Average Delivery Time per City (Top 20)

```python
delivery_city = (
    orders_full.groupby("customer_city")["delivery_time_days"]
    .mean()
    .sort_values()
    .head(20)
    .reset_index()
)

plt.figure(figsize=(10,6))
sns.barplot(data=delivery_city, x="delivery_time_days", y="customer_city")
plt.title("Top 20 Fastest Cities (Average Delivery Time)")
plt.xlabel("Delivery Time (days)")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

4.3 Freight Cost per City (Top 20 Highest)

```python
freight_city = (
    orders_full.groupby("customer_city")["freight_value"]
    .mean()
    .sort_values(ascending=False)
    .head(20)
    .reset_index()
)

plt.figure(figsize=(10,6))
sns.barplot(data=freight_city, x="freight_value", y="customer_city")
plt.title("Top 20 Cities with Highest Freight Cost")
plt.xlabel("Freight Value (BRL)")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```
