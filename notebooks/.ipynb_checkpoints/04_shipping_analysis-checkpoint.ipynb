ðŸŸ¦ 1. Load Libraries & Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../Data/engineered_dataset.csv")
```

ðŸŸ¦ 2. Delivery Time Overview

```python
df["delivery_time_days"].describe()
```

ðŸŸ¦ 3. Histogram Before Removing Outliers

```python
plt.figure(figsize=(10,5))
sns.histplot(df["delivery_time_days"], bins=100)
plt.title("Delivery Time Distribution (Before Cleaning)")
plt.show()
```

ðŸŸ¦ 4. Remove Outliers

```python
df_clean = df[(df["delivery_time_days"] >= 0) & (df["delivery_time_days"] <= 90)]
```

ðŸŸ¦ 5. Histogram After Cleaning

```python
plt.figure(figsize=(10,5))
sns.histplot(df_clean["delivery_time_days"], bins=40)
plt.title("Delivery Time Distribution (After Cleaning)")
plt.xlabel("Days")
plt.ylabel("Frequency")
plt.show()
```

ðŸŸ¦ 5. Histogram After Cleaning

```python
plt.figure(figsize=(10,5))
sns.histplot(df_clean["delivery_time_days"], bins=40)
plt.title("Delivery Time Distribution (After Cleaning)")
plt.xlabel("Days")
plt.ylabel("Frequency")
plt.show()
```

ðŸŸ¦ 6. Delivery Delay Analysis

```python
df_clean["delay_days"].describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_clean["delay_days"], bins=40)
plt.title("Delivery Delay Distribution")
plt.show()
```

ðŸŸ¦ 7. Shipping Time by State

```python
state_time = df_clean.groupby("customer_state")["delivery_time_days"].mean().sort_values()

plt.figure(figsize=(12,7))
sns.barplot(x=state_time.values, y=state_time.index)
plt.title("Average Delivery Time by State")
plt.xlabel("Delivery Time (days)")
plt.ylabel("State")
plt.show()
```

ðŸŸ¦ 8. Shipping Cost (Freight) Distribution

```python
plt.figure(figsize=(10,5))
sns.histplot(df_clean["freight_value"], bins=50)
plt.title("Freight Value Distribution")
plt.show()
```

ðŸŸ¦ 9. Shipping Cost vs Delivery Time

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=df_clean["delivery_time_days"],
    y=df_clean["freight_value"],
    alpha=0.3
)
plt.title("Shipping Cost vs Delivery Time")
plt.xlabel("Delivery Time (days)")
plt.ylabel("Freight Cost")
plt.show()
```

ðŸŸ¦ 10. Shipping Ratio (Freight / Product Price)

```python
plt.figure(figsize=(8,5))
sns.histplot(df_clean["shipping_ratio"], bins=50)
plt.title("Shipping Ratio Distribution")
plt.xlabel("Freight / Price Ratio")
plt.show()
```
