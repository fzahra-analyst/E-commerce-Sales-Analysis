ðŸŸ¦ 1. ØªØ­Ù…ÙŠÙ„ Ø§Ù„Ù…ÙƒØªØ¨Ø§Øª ÙˆØ§Ù„Ø¨ÙŠØ§Ù†Ø§Øª

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load engineered dataset
df = pd.read_csv("../Data/engineered_dataset.csv", parse_dates=["order_purchase_timestamp",
                                                              "order_approved_at",
                                                              "order_delivered_carrier_date",
                                                              "order_delivered_customer_date",
                                                              "order_estimated_delivery_date"])
```

ðŸŸ© 2. Ù…Ù„Ø®Ù‘Øµ Ø¹Ø§Ù… Ù„Ù„Ø¹Ù…Ù„Ø§Ø¡ (Customer Summary table)

```python
# Build customer summary (one row per customer)
customer_summary = df.groupby("customer_unique_id").agg(
    first_purchase=("order_purchase_timestamp", "min"),
    last_purchase=("order_purchase_timestamp", "max"),
    customer_order_count=("order_id", "nunique"),
    customer_total_spent=("price", "sum"),
    avg_order_value=("price", "mean")
).reset_index()

# Add recency in days relative to last date in dataset
snapshot_date = df["order_purchase_timestamp"].max() + pd.Timedelta(days=1)
customer_summary["recency_days"] = (snapshot_date - customer_summary["last_purchase"]).dt.days

# sort
customer_summary = customer_summary.sort_values("customer_total_spent", ascending=False)

# quick view
customer_summary.head()
```

ðŸŸ© 3. Ù†Ø¸Ø±Ø© Ø³Ø±ÙŠØ¹Ø©: ØªÙˆØ²ÙŠØ¹Ø§Øª Ùˆtop customers

```python
# Create a short version of the customer ID
customer_summary["customer_id_short"] = customer_summary["customer_unique_id"].str[:8]
```

```python
# Top 20 customers by total spent with short ID
top20 = customer_summary.nlargest(20, "customer_total_spent")

plt.figure(figsize=(10,6))
sns.barplot(
    data=top20,
    x="customer_total_spent",
    y="customer_id_short"
)

plt.title("Top 20 Customers by Total Spent")
plt.xlabel("Total Spent (BRL)")
plt.ylabel("Customer (Short ID)")
plt.tight_layout()
plt.show()
```

ðŸŸ© 4. Orders per customer distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(customer_summary["customer_order_count"], bins=20, log_scale=(False, True))
plt.title("Distribution of Orders per Customer")
plt.xlabel("Number of Orders")
plt.ylabel("Number of Customers (log scale)")
plt.show()
```

ðŸŸ© 5. Cohort / Retention (Ø¨Ø³ÙŠØ·) â€” Ø§Ø®ØªÙŠØ§Ø±ÙŠØŒ Ù„Ø¥Ø¸Ù‡Ø§Ø± Ø¥Ù…ÙƒØ§Ù†ÙŠØ© Ø§Ù„ØªØ­Ù„ÙŠÙ„ Ø§Ù„Ù…Ø³ØªÙ‚Ø¨Ù„ÙŠ

```python
# simple cohort by purchase month
df["purchase_month"] = df["order_purchase_timestamp"].dt.to_period("M")
cohort = df.groupby(["purchase_month", "customer_unique_id"])["order_id"].nunique().reset_index()
cohort_size = cohort.groupby("purchase_month")["customer_unique_id"].nunique()
cohort_size.head()
```

ðŸŸ© 6. RFM Analysis (Recency, Frequency, Monetary)

```python
import datetime as dt

# snapshot date already defined earlier
rfm = df.groupby("customer_unique_id").agg({
    "order_purchase_timestamp": lambda x: (snapshot_date - x.max()).days,
    "order_id": "nunique",
    "price": "sum"
}).reset_index()

rfm.columns = ["customer_unique_id", "Recency", "Frequency", "Monetary"]
rfm["Monetary"] = rfm["Monetary"].round(2)
rfm.head()
```

6.2 ØªÙ‚Ø·ÙŠØ¹ Ø§Ù„Ù€ R, F, M Ø¥Ù„Ù‰ Ø³ÙƒÙˆØ±Ø§Øª 1-5 (quintiles)

```python
# create R score (lower recency -> better -> higher score)
rfm["R_score"] = pd.qcut(rfm["Recency"], 5, labels=[5,4,3,2,1]).astype(int)
rfm["F_score"] = pd.qcut(rfm["Frequency"].rank(method="first"), 5, labels=[1,2,3,4,5]).astype(int)
rfm["M_score"] = pd.qcut(rfm["Monetary"], 5, labels=[1,2,3,4,5]).astype(int)

rfm["RFM_Score"] = rfm["R_score"].astype(str) + rfm["F_score"].astype(str) + rfm["M_score"].astype(str)
rfm.head()
```

6.3 ØªØ¹Ø±ÙŠÙ Segments Ø¨Ù†Ø§Ø¡Ù‹ Ø¹Ù„Ù‰ RFM (Ù‚Ø§Ù„Ø¨ Ø¨Ø³ÙŠØ·)

```python
def rfm_segment(row):
    r = row['R_score']
    f = row['F_score']
    m = row['M_score']
    if r >=4 and f >=4 and m >=4:
        return "Champions"
    if r >=3 and f >=3 and m >=3:
        return "Loyal"
    if r <=2 and f >=4:
        return "Promising"
    if r <=2 and f <=2 and m <=2:
        return "At Risk"
    return "Need Attention"

rfm['segment'] = rfm.apply(rfm_segment, axis=1)
rfm.segment.value_counts()
```

6.4 Ø¹Ø±Ø¶ Ù…ØªÙˆØ³Ø· Ø§Ù„Ù‚ÙŠÙ… Ù„ÙƒÙ„ Segment

```python
rfm_summary = rfm.groupby("segment").agg({
    "Recency":"mean",
    "Frequency":"mean",
    "Monetary":["mean","count"]
})
rfm_summary
```

ðŸŸ© 7. Visualize RFM segments

```python
plt.figure(figsize=(10,6))
sns.countplot(y="segment", data=rfm, order=rfm['segment'].value_counts().index)
plt.title("RFM Segment Counts")
plt.xlabel("Count")
plt.ylabel("Segment")
plt.show()
```

ðŸŸ© 8. Top segments (Monetary) â€” who to target

```python
top_segments = rfm.groupby("segment")["Monetary"].sum().sort_values(ascending=False)
top_segments.plot(kind="bar", figsize=(8,5))
plt.title("Total Monetary by RFM Segment")
plt.ylabel("Total Revenue")
plt.show()
```

ðŸŸ© 9. Export results (customers + rfm) for future use

```python
customer_summary.to_csv("../results/customer_summary.csv", index=False)
rfm.to_csv("../results/rfm_segments.csv", index=False)
```
