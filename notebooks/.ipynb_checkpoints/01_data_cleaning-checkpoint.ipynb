ðŸŸ¦ Load Libraries

```python
import pandas as pd
import numpy as np
```

ðŸŸ¦ Load All CSV Files

```python
customers = pd.read_csv("../Data/olist_customers_dataset.csv")
orders = pd.read_csv("../Data/olist_orders_dataset.csv")
order_items = pd.read_csv("../Data/olist_order_items_dataset.csv")
products = pd.read_csv("../Data/olist_products_dataset.csv")
sellers = pd.read_csv("../Data/olist_sellers_dataset.csv")
geolocation = pd.read_csv("../Data/olist_geolocation_dataset.csv")
payments = pd.read_csv("../Data/olist_order_payments_dataset.csv")
reviews = pd.read_csv("../Data/olist_order_reviews_dataset.csv")
```

ðŸŸ¦ Merge Datasets into One Master Table

```python
df = orders.merge(customers, on="customer_id", how="left") \
           .merge(order_items, on="order_id", how="left") \
           .merge(products, on="product_id", how="left") \
           .merge(sellers, on="seller_id", how="left") \
           .merge(payments, on="order_id", how="left") \
           .merge(reviews, on="order_id", how="left")
```

ðŸŸ¦ Convert Dates

```python
date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col])
```

ðŸŸ¦ Handle Missing Values

```python
df["product_weight_g"].fillna(df["product_weight_g"].median(), inplace=True)
df["product_volume_cm3"] = (
    df["product_length_cm"] * df["product_height_cm"] * df["product_width_cm"]
)

df["order_delivered_customer_date"].fillna(df["order_estimated_delivery_date"], inplace=True)
```

ðŸŸ¦ Reset Index + Basic Cleaning

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
```

ðŸŸ¦ Save Clean Data For Next Steps

```python
df.to_csv("../Data/cleaned_dataset.csv", index=False)
```
